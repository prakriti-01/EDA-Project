##  Data Cleaning

```python
df.info()
```

### check for null Values 

```python
df.isnull().sum()
```

### deleting columns,Description and URL since they is no need for them 

```python
df.drop('URL', inplace=True, axis=1)
```

```python
df.drop('Description', inplace=True, axis=1)
```

```python
df['DiscountOffer'].nunique()
```

```python
df.drop('DiscountOffer', inplace=True, axis=1)
```

```python
df.head()
```

```python
# Filling tye null values 
```

```python
# fillimg the Ratings column with 'No ratings' since no one has given any ratings on the product
df["Ratings"].fillna("No Ratings", inplace = True)
```

```python
df["Reviews"].fillna("No Reviews", inplace = True)

```

```python
# Making a new columns as 'DiscountPerc' and cslculating the discount Percentage in it,since many values were missing 
df['DiscountPerc'] = round((df['OriginalPrice (in Rs)']-df['DiscountPrice (in Rs)'])*100/df['OriginalPrice (in Rs)'])
df.head()
```

```python
df.isnull().sum()
```

```python
# filling the 'DiscountPerc' column as 0 where there is no discount
df['DiscountPerc'] = df['DiscountPerc'].fillna(0)
```

```python
# Filling the 'DiscountPrice' as 'OriginalPrice' where 'DiscountPerc'= 0, since there is no discount in the product
df['DiscountPrice (in Rs)'] = np.where(df['DiscountPerc']==0.0,df['OriginalPrice (in Rs)'],df['DiscountPrice (in Rs)'])
```
